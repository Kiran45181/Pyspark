```python
import pandas as pd

# Load CSVs
sales_df = pd.read_csv("sales_data.csv")
customer_df = pd.read_csv("customer_data.csv")
product_df = pd.read_csv("product_data.csv")

# Display all rows (optional)
pd.set_option('display.max_rows', None)

# Show them
print("Sales Data:")
display(sales_df)

print("Customer Data:")
display(customer_df)

print("Product Data:")
display(product_df)
```

```python
print("Sales Data shape:", sales_df.shape)
print("Customer Data shape:", customer_df.shape)
print("Product Data shape:", product_df.shape)
```

##Section 1: Aggregations

##1.Total Revenue Per Product

```python
# Add revenue column
sales_df['revenue'] = sales_df['quantity'] * sales_df['price']

# Total revenue per product
total_revenue_per_product = sales_df.groupby('product_id')['revenue'].sum().reset_index()

# Merge with product names for clarity
total_revenue_per_product = total_revenue_per_product.merge(product_df, on='product_id', how='left')

print("Total Revenue Per Product:")
display(total_revenue_per_product)
```

##2. Total Quantity Sold Per Customer

```python
total_quantity_per_customer = sales_df.groupby('customer_id')['quantity'].sum().reset_index()

# Merge with customer names
total_quantity_per_customer = total_quantity_per_customer.merge(customer_df, on='customer_id', how='left')

print("Total Quantity Sold Per Customer:")
display(total_quantity_per_customer)
```

##3. Average Revenue Per Customer

```python
avg_revenue_per_customer = sales_df.groupby('customer_id')['revenue'].mean().reset_index()

# Merge with customer info
avg_revenue_per_customer = avg_revenue_per_customer.merge(customer_df, on='customer_id', how='left')

print("Average Revenue Per Customer:")
display(avg_revenue_per_customer)
```

##4. Monthly Sales Total

```python
# Convert sale_date to datetime
sales_df['sale_date'] = pd.to_datetime(sales_df['sale_date'])

# Extract month
sales_df['month'] = sales_df['sale_date'].dt.to_period('M')

# Total sales per month
monthly_sales_total = sales_df.groupby('month')['revenue'].sum().reset_index()

print("Monthly Sales Total:")
display(monthly_sales_total)
```

##5. Count of Sales Per Category


```python
# Join sales with product categories
sales_product = sales_df.merge(product_df, on='product_id', how='left')

# Count sales per category
count_sales_per_category = sales_product.groupby('category')['sale_id'].count().reset_index(name='sales_count')

print("Count of Sales Per Category:")
display(count_sales_per_category)
```

##6. Top 3 Most Expensive Products Sold

```python
top_3_products = total_revenue_per_product.sort_values(by='revenue', ascending=False).head(3)

print("Top 3 Most Expensive Products (by revenue):")
display(top_3_products)
```

##Section 2 : JOINS

##1. Join Sales with Product Information

```python
# Join sales with product info
sales_product_join = sales_df.merge(product_df, on='product_id', how='inner')

print("Sales + Product Info:")
display(sales_product_join.head())
```

##2. Join Sales with Customer Information

```python
# Join sales with customer info
sales_customer_join = sales_df.merge(customer_df, on='customer_id', how='inner')

print("Sales + Customer Info:")
display(sales_customer_join.head())
```

##3. Inner Join (Sales + Product) for Specific Category (e.g., Gadgets)


```python
# Inner join sales & product, filter for Gadgets
gadgets_sales = sales_df.merge(product_df, on='product_id', how='inner')
gadgets_sales = gadgets_sales[gadgets_sales['category'] == 'Gadgets']

print("Gadgets Category Sales (Inner Join):")
display(gadgets_sales.head())
```

##4. Left Join Between Sales and Products

```python
left_join_sales_product = sales_df.merge(product_df, on='product_id', how='left')

print("Left Join (Sales + Product):")
display(left_join_sales_product.head())
```

##5. Self-Join Sales Table

```python
self_join_sales = sales_df.merge(
    sales_df,
    on='product_id',
    suffixes=('_first', '_second')
)

# Example: Different dates comparison
self_join_sales = self_join_sales[self_join_sales['sale_date_first'] != self_join_sales['sale_date_second']]

print("Self-Join Sales (Compare different dates for same product):")
display(self_join_sales.head())
```

##6. Full Outer Join Sales and Products


```python
full_outer_join = pd.merge(
    sales_df,
    product_df,
    on='product_id',
    how='outer',
    indicator=True
)

print("Full Outer Join (Sales + Products):")
display(full_outer_join.head())
```

##7. Join Multiple Tables (Sales + Customer + Product)

```python
# Sales + Product
sales_product = sales_df.merge(product_df, on='product_id', how='inner')

# Add Customer info
sales_full = sales_product.merge(customer_df, on='customer_id', how='inner')

print("Sales + Product + Customer Info:")
display(sales_full.head())
```

**Section 3: Additional Transformations**

##1. Filter Data for Specific Date Range

```python
filtered_sales = sales_df[
    (sales_df['sale_date'] >= '2025-07-10') &
    (sales_df['sale_date'] <= '2025-07-15')
]

print("Sales between 2025-07-10 and 2025-07-15:")
display(filtered_sales)
```

##2. Top 5 Customers by Total Spend

```python
# Total spend = sum of revenue
top_5_customers = sales_df.groupby('customer_id')['revenue'].sum().reset_index()

# Sort and get top 5
top_5_customers = top_5_customers.sort_values(by='revenue', ascending=False).head(5)

# Merge with customer names
top_5_customers = top_5_customers.merge(customer_df, on='customer_id', how='left')

print("Top 5 Customers by Spend:")
display(top_5_customers)
```

##3. Categorize Customers Based on Spend

```python
# Categorize
def categorize_spend(amount):
    if amount < 100:
        return "Low"
    elif amount <= 300:
        return "Medium"
    else:
        return "High"

customer_spend = sales_df.groupby('customer_id')['revenue'].sum().reset_index()
customer_spend['spend_category'] = customer_spend['revenue'].apply(categorize_spend)

# Merge with customer info
customer_spend = customer_spend.merge(customer_df, on='customer_id', how='left')

print("Customers Categorized by Spend:")
display(customer_spend)
```

##4. First and Last Purchase Per Customer

```python
# Group by customer and get min/max dates
purchase_dates = sales_df.groupby('customer_id')['sale_date'].agg(['min', 'max']).reset_index()

# Merge with customer info
purchase_dates = purchase_dates.merge(customer_df, on='customer_id', how='left')

print("First and Last Purchase Dates per Customer:")
display(purchase_dates)
```

##5. Customer Churn (No Purchase in Last 30 Days)

```python
import pandas as pd

# Set current date
current_date = pd.to_datetime('2025-07-28')

# Last purchase per customer
last_purchase = sales_df.groupby('customer_id')['sale_date'].max().reset_index()

# Convert to datetime
last_purchase['sale_date'] = pd.to_datetime(last_purchase['sale_date'])

# Check churn (no purchase in last 30 days)
last_purchase['days_since_last_purchase'] = (current_date - last_purchase['sale_date']).dt.days
churn_customers = last_purchase[last_purchase['days_since_last_purchase'] > 30]

# Merge with customer info
churn_customers = churn_customers.merge(customer_df, on='customer_id', how='left')

print("Customers with No Purchase in Last 30 Days:")
display(churn_customers)
```
